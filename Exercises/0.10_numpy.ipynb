## 0. Dice simulations (*)

Simulations of one die,  

&nbsp; a) Calculate theoretical mean of a dice (six-sided) (*)

&nbsp; b) Now do simulations of

- 10
- 100
- 1000
- 10000
- 100000
- 1000000
- 10000000

number of dices and plot their means. (*)


```python
import random
import matplotlib.pyplot as plt

# Funktion för att simulera tärningskast
def simulate_dice_rolls(n):
    rolls = [random.randint(1, 6) for _ in range(n)]
    return sum(rolls) / n

# Lista för att hålla antalet kast och deras medelvärden
num_rolls = [10, 100, 1000, 10000, 100000, 1000000, 10000000]
means = []

# Utför simuleringar och spara medelvärden
for n in num_rolls:
    mean = simulate_dice_rolls(n)
    means.append(mean)


plt.plot(num_rolls, means, marker='o')
plt.xscale('log')  # Använd logaritmisk skala på x-axeln för bättre läsbarhet
plt.xlabel('Antal tärningskast')
plt.ylabel('Beräknat medelvärde')
plt.title('Medelvärde av tärningskast vid olika antal simuleringar')
plt.grid(True)
plt.show()
```
